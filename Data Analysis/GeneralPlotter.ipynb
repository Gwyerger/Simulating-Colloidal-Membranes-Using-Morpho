```python

for i in range(0,1):
    gam = 0.5 + i*0.1

    for j in range(0,1):
        ff = 0.6 + j*0.1

        for k in range(0,11):
            q =  1 + k*0.1

            gam = round(gam, ndigits = 2)
            ff = round(ff, ndigits = 2)
            q = round(q, ndigits = 2)
            if k==0 or k==10: q = round(q)

            name1 = "EnergiesItg"+str(gam)+"f"+str(ff)+"q"+str(q)+"sopt.txt"
            name2 = "EnergiesItg"+str(gam)+"f"+str(ff)+"q"+str(q)+"fopt.txt"
            imgname = "SinglePore"+str(i)+"f"+str(j)+"q"+str(k)+".png"
            
            file1 = open(path+name1, "r")

            snapshot = img.imread(pathimg+imgname)

            strdata1 = file1.readlines()
            for ii in range(0, len(strdata1)):
                strdata1[ii] = float(strdata1[ii])
            datarray1 = np.array(strdata1)
            #Now Grabbing second data file:
            file2 = open(path+name2, "r")
            strdata2 = file2.readlines()
            for ii in range(0, len(strdata2)):
                strdata2[ii] = float(strdata2[ii])
            datarray2 = np.array(strdata2)
            fig,ax = plt.subplots(3)
            ax[0].set_title('Energy over Iterations in Morpho for $\gamma=$'+str(gam)+", $P=$"+str(ff)+", $q=$"+str(q), fontsize=16)
            ax[0].plot(np.arange(0, len(datarray1), 1), datarray1[:], label = "sopt")
            ax[1].plot(np.arange(0, len(datarray2), 1), datarray2[:], label = "fopt")

            ax[0].set_ylabel("Energy")
            ax[0].set_xlabel("Iterations")
            ax[0].set_ylim((datarray1.min(),datarray1.max()))
            ax[1].set_ylabel("Energy")
            ax[1].set_xlabel("Iterations")
            ax[1].set_ylim((datarray2.min(),datarray2.max()))
            ax[0].grid(visible=True, which = 'major')
            ax[0].legend()
            ax[1].grid(visible=True, which = 'major')
            ax[1].legend()
            ax[2].imshow(snapshot)
            fig.set_size_inches(10, 12)
            # fig.savefig(path2 +"EnergiesImgg"+str(gam)+"f"+str(ff)+"q"+str(q)+".png", dpi = 300)
```

```python
pop = "\n"
delimiter = ","
poresizes = open(pathPSD+"PoreSizeData.txt","r")
sizedata = poresizes.readlines()[1:]
dict = {"g":[], "f":[], "q":[], "ps":[]}
for line in sizedata:
    line = line.replace(pop, "")
    line = line.split(delimiter)
    dict["g"].append(float(line[0]))
    dict["f"].append(float(line[1]))
    dict["q"].append(float(line[2]))
    dict["ps"].append(float(line[3]))
df = pnd.DataFrame(dict)

f1slice = df["f"] == 0.6
f1sliced = df[f1slice]
f2slice = df["f"] == 0.7
f2sliced = df[f2slice]
f3slice = df["f"] == 0.8
f3sliced = df[f3slice]

gs = list(set(df["g"]))
qs = list(set(df["q"]))

gs.sort()
qs.sort()

print(f1sliced)

arr = [[f1sliced[(f1sliced["q"]==q) & (f1sliced["g"]==g)]["ps"].iloc[0] for q in qs] for g in gs]
arr = np.array(arr)
print(arr)

ax = sns.heatmap(arr, linewidth=0.5)
plt.show()
```
