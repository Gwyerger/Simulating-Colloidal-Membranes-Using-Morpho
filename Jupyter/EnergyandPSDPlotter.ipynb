```python
pop = "\n"
delimiter = ","
poresizes = open(pathPSD+"PoreSizeData.txt","r")
sizedata = poresizes.readlines()[1:]
dict = {"g":[], "f":[], "q":[], "ps":[]}
for line in sizedata:
    line = line.replace(pop, "")
    line = line.split(delimiter)
    dict["g"].append(float(line[0]))
    dict["f"].append(float(line[1]))
    dict["q"].append(float(line[2]))
    dict["ps"].append(float(line[3]))
df = pnd.DataFrame(dict)

f1slice = df["f"] == f1
f1sliced = df[f1slice]
# f2slice = df["f"] == 0.7
# f2sliced = df[f2slice]
# f3slice = df["f"] == 0.8
# f3sliced = df[f3slice]

fs = list(set(df["f"]))
gs = list(set(df["g"]))
qs = list(set(df["q"]))

fs.sort()
gs.sort()
qs.sort()

arr1 = np.array([[f1sliced[(f1sliced["q"]==q) & (f1sliced["g"]==g)]["ps"].iloc[0] for q in qs] for g in gs])

qs = np.around(qs/np.sqrt(2), decimals=2)
gs = np.around(gs/np.sqrt(2), decimals=2)

# Units of lambda^2, where lambda = 1/sqrt2

arr1 = np.array([[round(arr1[i, j]*2, ndigits=2) if arr1[i, j] > 0 else 0 for j in np.arange(0, len(arr1[1, :]), 1)] for i in np.arange(0, len(arr1[:, 1]), 1)])

# arr2 = [[f2sliced[(f2sliced["q"]==q) & (f2sliced["g"]==g)]["ps"].iloc[0] for q in qs] for g in gs]
# arr2 = np.array(arr2)*4

# arr3 = [[f3sliced[(f3sliced["q"]==q) & (f3sliced["g"]==g)]["ps"].iloc[0] for q in qs] for g in gs]
# arr3 = np.array(arr3)*4

# vmin = 0, vmax = 30

fig = plt.figure(constrained_layout=True)
fig.suptitle("p$^*$ = "+str(fs[0]/2), fontsize=26)
gspec = fig.add_gridspec(1,4, width_ratios= [1, 26, 2, 1])
ax1 = fig.add_subplot(gspec[0,0],  facecolor='white')
ax1.tick_params(bottom=False, left=False, labelbottom=False, labelleft=False)
ax2 = fig.add_subplot(gspec[0,1])
ax3 = fig.add_subplot(gspec[0,2])
ax4 = fig.add_subplot(gspec[0,3],  facecolor='white')
ax4.tick_params(bottom=False, left=False, labelbottom=False, labelleft=False)


fig.set_size_inches(14, 8)
sns.heatmap(arr1, annot = False, linewidth=0.5, linecolor="white",  cbar_ax=ax3,  vmin = 0, vmax = 30, ax = ax2)
ax2.invert_yaxis()
ax2.set_xlabel("q$^*$", fontsize = 24)
ax1.set_ylabel("$\gamma^*$", fontsize = 24, rotation = 0)
ax4.set_ylabel("Pore Area (units of $\lambda_t^2$)", fontsize = 24)
ax4.yaxis.set_label_position("right")
ax2.set_yticks(np.linspace(0.5, 9.5, 7), labels=np.around(np.linspace(0.1, 0.7, 7),decimals=2), rotation = 0)
ax2.set_xticks(np.linspace(0.5, 19.5, 14), labels=np.around(np.linspace(0.0, 1.3, 14),decimals=2))

plt.show()
fig.savefig(pathPSD +f"PSD P={fs[0]/2}.png", dpi = 400)

```
