$\text{Create and write to the new morpho files}$

```python
def overwrite(parentfile, childfile, q, g):
            with open(parentfile, 'r') as file:
                # read a list of lines into data
                data = file.readlines()

            data[17] = f"var q = {q} \n"
            data[18] = f"var gam = {g} \n"
            data[30] = f'var pathexport = \"{childpathlinux}/q={q} g={g}\"\n'
            # and write everything back
            with open(childfile, 'w') as newfile:
                newfile.writelines( data )

for q in qchirality:
    for g in glinetension:
        childfile = childpath + f'\\q={q} g={g}' + r"\SinglePoreR6.morpho"
        parentfile = parentpath+r"\SinglePoreR6Main.morpho"
        overwrite(parentfile, childfile, q, g)
```

$\text{Call the new files in WSL terminal}$

```python
print(f"Starting batch of SinglePoreR6.morpho \n Batch size: {len(qchirality)} x {len(glinetension)} \n qs = {qchirality} \n gs = {glinetension} \n")
```

```python
ti = time.time()
for q in qchirality:
    for g in glinetension:
        t1 = time.time()
        file = childpathlinux + f'/q={q} g={g}' + r"/SinglePoreR6.morpho"
        call = subprocess.call(['wsl', 'morpho5', file])
        t2 = time.time()
        print(f"call: {call} q = {q}, g = {g}, time of simulation = {(t2 - t1)/60} min \n")
tf = time.time()
print(f"time of batch: {(tf - ti)/60} min")
```
