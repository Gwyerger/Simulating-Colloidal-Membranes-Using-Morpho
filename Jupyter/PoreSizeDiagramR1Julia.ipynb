# Creating a Phase Diagram in Julia

```julia
# We devise a PVPython script that tells us the size of a pore.
# The data we receive is in a text document of the form bool.

path = "C:\\Users\\gabey\\OneDrive\\Desktop\\Fall 2023\\SM Research\\Phase Diagram VTK Exports and PP\\Single Pore Round 1\\"
name1 = "PoreSizeData.txt"
delimiter = ","
pop = "\n"
file = open(path*name1, "r")
strdata = readlines(file)


gs = []
qs = []

for line in strdata[begin+1:1:end]
    line = replace(line, pop => "")
    line = split(line, delimiter)
    g = parse(Float64, line[1])
    push!(gs,g)
    q = parse(Float64, line[2])
    push!(qs,q)
end

gs = union(gs)
qs = union(qs)
sizes = rand(length(qs), length(gs))


for i in 1:length(qs)
    for j in 1:(length(gs))
        line = strdata[i+(j-1)*length(qs)+1]
        line = replace(line, pop => "")
        line = split(line, delimiter)
        sizes[i, j] = parse(Float64, line[3])
    end
end
```

```julia
plt = heatmap(qs, gs, sizes', aspect_ratio = 1, color = cgrad(:devon, rev=false))
plot!(plt,title = "Pore Size Diagram",minorgrid=true,  xlabel="qλₜ", ylabel="γλₜ/K", ylims = [gs[begin]-0.5, gs[end]+0.5], xticks = qs*0.1, yticks=gs*0.1, colorbar_title=join(["Pore Size (μm)", L"^2"]))
```
